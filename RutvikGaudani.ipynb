```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url="https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv"
s_data = pd.read_csv(url)
s_data.head(10)
```

```python
s_data.plot(x='Hours', y='Scores', style='o')
plt.title('Hours_study v/s Percentage Score')
plt.xlabel('Hours Study')
plt.ylabel('Percentage Score')
plt.show()
```

```python
x = s_data.iloc[: , :-1].values
y = s_data.iloc[: , 1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
Regressor1 = LinearRegression()
Regressor1.fit(x,y)
print("Training Model")
```

```python
line = Regressor1.coef_*x+Regressor1.intercept_
```

```python
plt.scatter(x,y)
plt.plot(x,line)
plt.show()
```

```python
print(x_test)
y_pred = Regressor1.predict(x_test)
```

```python
df = pd.DataFrame({'Actual':y_test, 'Predicted':y_pred})
df
```

```python
print("Training Score", Regressor1.score(x_train, y_train))
print("Testing Score", Regressor1.score(x_test, y_test))
```

```python
df.plot(kind='bar', figsize=(7,7))
plt.show()
```

```python
hours=9.25
test = np.array([hours])
test = test.reshape(-1,1)
own_pred = Regressor1.predict(test)
print("no of hours = {}".format(hours))
print("predicted score = {}".format(own_pred[0]))
```

```python
import numpy as np
from sklearn import metrics
print('Mean absolute error: ', metrics.mean_absolute_error(y_test,y_pred))
print('Mean squared error: ', metrics.mean_squared_error(y_test,y_pred))
print('Root Mean squared error: ', np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
print('Explainrd variance score: ', metrics.explained_variance_score(y_test,y_pred))
```
